```python
all_month = pd.read_csv('./all_month.csv')
all_month.head()
```

#### Clean up the date/Drop NaN rows

#### find 'Or' and delete it

nan_df = all_month[all_month.isna().any(axis=1)] - вычисляем количество строк с NaN

```python
nan_df = all_month[all_month.isna().any(axis=1)]
nan_df
```

```python
all_month = all_month.dropna(how='all')
all_month
```

```python
all_month = all_month[all_month['Order Date'].str[0:2] != 'Or']
```

```python
all_month['Quantity Ordered'] = pd.to_numeric(all_month['Quantity Ordered'])
all_month['Price Each'] =  pd.to_numeric(all_month['Price Each'])
```

#### what is the best month for sales?

##### augment (увеличить),количество столбцов

all_month['Month']=3 добавляет столбец
all_month['Month'] = all_month['Order Date'].str[0:2] - созд стоб Month,перевел знач столбца Order Date в строку,взял из нее первые два символа и записал в столбец Month

```python
all_month['Month'] = all_month['Order Date'].str[0:2]
all_month['Month'] = all_month['Month'].astype('int32')
all_month
```

```python
all_month['Sales'] = all_month['Quantity Ordered']*all_month['Price Each']
all_month
```

#### How to rnow the biggest sales on the sity name

##### add a new colomn "Sity".Используя метод apply()

Метод apply() — это инструмент для преобразования объекта DataFrame . аpply() можно применять как к одному столбцу, так и к нескольким.lambda x: x.split(',')[1] - анонимная функция,разделил строку по запятой и взял второй элемент

all_month['City'] = all_month['Purchase Address'].apply(lambda x:f"{ x.split(',')[1]} ({get_state(x)})") запись с использованием f строки

```python
def get_state(addres):
    return addres.split(',')[2].split(' ')[1]


all_month['City'] = all_month['Purchase Address'].apply(lambda x: x.split(',')[1] + ' ('+ get_state(x)+')')
all_month
```

```python
results = all_month.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt
month = range(1,13)
plt.bar(month, results['Sales'])
plt.ylabel('Sales in USA($)')
plt.xlabel('Month')
plt.show()

```

```python
results_city = all_month.groupby('City').sum()
results_city
```

```python
import matplotlib.pyplot as plt

cities = [city for city, df in all_month.groupby('City')]
plt.bar(cities, results_city['Sales'])
plt.xticks(cities, rotation='vertical', size=8)

plt.ylabel('Sales in USA($)')
plt.xlabel('City name')
plt.show()
```
